# 📌 Avaliador de Argumentos Lógicos
Este programa tem como objetivo analisar argumentos da lógica proposicional a partir de hipóteses e uma conclusão fornecidas pelo usuário.

A partir da entrada das proposições, o código realiza automaticamente:

* a leitura e organização das hipóteses e da conclusão do argumento;
* a construção de uma expressão lógica que representa o argumento no formato (hipóteses → conclusão);
* a identificação das letras de proposição presentes na expressão;
a geração de todas as combinações possíveis de valores lógicos (True e False);
* a construção da tabela verdade correspondente parcial;
* a avaliação lógica do argumento;
* a verificação de sua validade.


### 🧠 Como utilizar

1. Informe quantas hipóteses o argumento possui (sem contar a conclusão).

2. Insira cada hipótese utilizando os conectivos:
    * ∧ (conjunção)
    * ∨ (disjunção)
    * → (condicional)
    * ↔ (bicondicional)
    * ¬ (negação)

3. Insira a conclusão do argumento.


4. O programa irá construir automaticamente a expressão lógica completa do argumento e gerar a tabela verdade parcial correspondente

5. Ao final, será possível verificar se o argumento é:

    * Válido → quando a implicação (hipóteses → conclusão) é sempre verdadeira
    * Inválido → quando existe pelo menos um caso em que as hipóteses são verdadeiras e a conclusão é falsa


**O usuário pode optar por analisar múltiplos argumentos em sequência.**

### ⚠️ Observações
Utilize apenas letras para representar proposições (ex: p, q, r).
Parênteses podem ser usados para definir a ordem das operações.
Certifique-se de utilizar corretamente os conectivos lógicos indicados.

### ▶️ Como iniciar o programa
Para iniciar o uso do avaliador, basta executar a célula de código abaixo.
O programa será iniciado automaticamente, solicitando ao usuário a inserção das proposições (hipóteses e conclusão) e guiando todo o processo de análise.
<br>

Caso esteja utilizando um notebook (Jupyter ou Colab), execute a célula pressionando: `Shift + Enter`

Após isso, siga as instruções exibidas no terminal para utilizar o programa.

```python
while True:

    print("Os conectivos disponíveis para uso são '∧', '∨', '→', '↔' e '¬'\nVocê pode copiá-los da frase acima")
    qtd_hipoteses = int(input("Insira quantas hipóteses o argumento possui (sem contar a conclusão):"))
    argumento = str()
    assert qtd_hipoteses > 0, "A quantidade de hipóteses deve ser maior que 0"

    for i in range(qtd_hipoteses):
        hipotese = input(f"Insira a hipótese {i+1} do argumento a ser analisado:")
        print(f"hipótese {i+1}: {hipotese}")
        if i == 0:
            argumento = argumento + hipotese
        else:
            argumento = argumento + "∧" + hipotese
    conclusao = input("Insira a conclusão do argumento a ser analisado:")
    print(f"conclusão: {conclusao}")
    argumento = argumento + "→" + conclusao
    separados = []
    letras = set()

    for i in argumento:
        if i in ['∧', '∨', '→', '↔', '(', ')', '¬']:
            separados.append(i)
        elif i.isalpha():
            separados.append(i)
            letras.add(i)

    letras = sorted(list(letras))

    def gerar_combinacoes(letras, atual=[], resultado=[]):
        if len(atual) == len(letras):
            resultado.append(atual.copy())
            return resultado

        gerar_combinacoes(letras, atual + [True], resultado)
        gerar_combinacoes(letras, atual + [False], resultado)

        return resultado

    combinacoes = gerar_combinacoes(letras)

    dados = {}

    for letra in letras:
        dados[letra] = []
        
    for linha in combinacoes:
        for i in range(len(letras)):
            dados[letras[i]].append(linha[i])

    def converte_rpn(separados):
        saida = []
        pilha = []

        precedencia = {
            '¬': 3,
            '∧': 2,
            '∨': 1,
            '→': 0,
            '↔': 0
        }

        for caractere in separados:

            if caractere.isalpha():
                saida.append(caractere)

            elif caractere == '(':
                pilha.append(caractere)

            elif caractere == ')':
                while pilha and pilha[-1] != '(':
                    saida.append(pilha.pop())
                pilha.pop()

            else:
                while ((pilha and pilha[-1] != '(') and (precedencia.get(pilha[-1], 0) >= precedencia.get(caractere, 0))):
                    saida.append(pilha.pop())

                pilha.append(caractere)

        while pilha:
            saida.append(pilha.pop())

        return saida

    rpn = converte_rpn(separados)

    def conjuncao(v1,v2):
        resultado = (v1 and v2)
        return resultado

    def disjuncao(v1,v2):
        resultado = (v1 or v2)
        return resultado

    def negacao(v):
        resultado = (not v)
        return resultado

    def condicional(v1,v2):
        resultado = (not(v1) or v2)
        return resultado

    def bicondicional(v1,v2):
        resultado = (condicional(v1,v2) and condicional(v2,v1))
        return resultado

    def avalia_rpn(rpn, valores):
        pilha = []

        for caractere in rpn:

            if caractere.isalpha():
                pilha.append(valores[caractere])

            elif caractere == '¬':
                v = pilha.pop()
                pilha.append(negacao(v))

            else:
                v2 = pilha.pop()
                v1 = pilha.pop()

                if caractere == '∧':
                    pilha.append(conjuncao(v1,v2))

                elif caractere == '∨':
                    pilha.append(disjuncao(v1,v2))

                elif caractere == '→':
                    pilha.append(condicional(v1,v2))

                elif caractere == '↔':
                    pilha.append(bicondicional(v1,v2))

        return pilha[0]

    dados[argumento] = []
    for linha in combinacoes:
        valores = dict(zip(letras, linha))
        resultado = avalia_rpn(rpn, valores)
        dados[argumento].append(resultado)

    import pandas as pd

    tabela = pd.DataFrame(dados)
    print(tabela) 

    if tabela.iloc[:, -1].all():
        print(f"O argumento inserido é válido pois é uma tautologia")
    else:
        print(f"O argumento inserido não é válido pois não é uma tautologia")


    quer_repetir = input("Gostaria de analisar um outro argumento? (s/n)")
    if quer_repetir.lower() == "s":
            pass

    elif quer_repetir.lower() == "n":
        print("Programa encerrado. Quando quiser reiniciar, rode o código novamente")
        break

    else:
        print("Não entendi sua resposta, tente novamente")
```
