#### **3.** Posteriormente, foi criado um novo código, implementando:
* função input para o usuário inserir a proposição a ser analisada (com apenas 2 letras de proposição e 1 conectivo)
* separando a string inserida em uma lista com os componentes separados
* definindo funções para traduzir cada conectivo (dado em símbolos da lógica formal) para a linguagem do python

#### O código dessa etapa foi inserido a seguir:

```
import pandas as pd

proposicao = input("Insira a proposição a ser analisada:")
separados = proposicao.split()

def conjuncao(v1,v2):
    resultado = (v1 and v2)
    return resultado

def disjuncao(v1,v2):
    resultado = (v1 or v2)
    return resultado

def negacao(v):
    resultado = (not v)
    return resultado

def condicional(v1,v2):
    resultado = (not(v1) or v2)
    return resultado

def bicondicional(v1,v2):
    resultado = (condicional(v1,v2) and condicional(v2,v1))
    return resultado

def conectivo(v1, v2, conectivo):
    if conectivo == "∧":
        return conjuncao(v1,v2)
    elif conectivo == "∨":
        return disjuncao(v1,v2)
    elif conectivo == "→":
        return condicional(v1,v2)
    elif conectivo == "↔":
        return bicondicional(v1,v2)


dados = {
    separados[0]: [],
    separados[2]: [],
    proposicao: []
}
        
valores = [True, False]
for p in valores:
    for q in valores:
        dados[separados[0]].append(p)
        dados[separados[2]].append(q)
        dados[proposicao].append(conectivo(p, q, separados[1]))
tabela = pd.DataFrame(dados)
print(tabela)    
```

##### **Parte 1:** gerando as primeiras colunas da tabela verdade (apenas com as letras de proposição)
```
def gerar_combinacoes(letras, atual=[], resultado=[]):
    if len(atual) == len(letras):
        resultado.append(atual.copy())
        return

    gerar_combinacoes(letras, atual + [True], resultado)
    gerar_combinacoes(letras, atual + [False], resultado)

    return resultado

combinacoes = gerar_combinacoes(letras)

dados = {}

for letra in range(len(letras)):
    dados[letra] = []

for linha in combinacoes:
    for i in range(len(letras)):
        dados[letras[i]].append(linha[i])
```
