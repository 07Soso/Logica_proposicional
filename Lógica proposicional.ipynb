```python
import pandas as pd

print("Os conectivos disponíveis para uso são '∧', '∨', '→' e '↔'\nVocê pode copiá-los da frase acima")
proposicao = input("Insira a proposição a ser analisada:")
separados = proposicao.split()

assert len(separados) == 3, "A proposição deve conter 3 elementos: letra de proposição, conectivo, letra de proposição - nessa ordem"
assert separados[0].isupper(), "As letras de proposição devem ser maiúsculas"
assert separados[2].isupper(), "As letras de proposição devem ser maiúsculas"
assert separados[1] == ("∧" or "∨" or "→" or "↔"), "O conectivo inserido não é válido, escolha entre '∧', '∨', '→' ou '↔'"

def conjuncao(v1,v2):
    resultado = (v1 and v2)
    return resultado

def disjuncao(v1,v2):
    resultado = (v1 or v2)
    return resultado

def negacao(v):
    resultado = (not v)
    return resultado

def condicional(v1,v2):
    resultado = (not(v1) or v2)
    return resultado

def bicondicional(v1,v2):
    resultado = (condicional(v1,v2) and condicional(v2,v1))
    return resultado

    

def conectivo(v1, v2, conectivo):
    if conectivo == "∧":
        return conjuncao(v1,v2)
    elif conectivo == "∨":
        return disjuncao(v1,v2)
    elif conectivo == "→":
        return condicional(v1,v2)
    elif conectivo == "↔":
        return bicondicional(v1,v2)


dados = {
    separados[0]: [],
    separados[2]: [],
    proposicao: []
}
        
valores = [True, False]
for p in valores:
    for q in valores:
        dados[separados[0]].append(p)
        dados[separados[2]].append(q)
        dados[proposicao].append(conectivo(p, q, separados[1]))
tabela = pd.DataFrame(dados)
print(tabela)    
```
